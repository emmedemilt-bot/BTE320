```python
def loan_payment(principal,annual_rate, months):
  assert isinstance(principal, (int,float))
  assert isinstance(annual_rate,(int,float))
  assert isinstance(months,int)

  if principal <= 0 or annual_rate < 0 or months <= 0:
    raise ValueError("invalid input values")

  try:
    r = annual_rate / 12
    if r ==0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1+r) ** (-months))
  except Exception as e:
      raise Exception(f"Error: {e}")

  return pmt
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_courses(self):
    return self.__courses

  def get_name(self):
    return self.__name

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self,course):
    if course in self.__courses:
      print(f"Student {self.__name} has already enrolled in course {course}.")
    else:
      self.__courses.append(course)

```

```python
s = Student("Jane Doe", 12345)
print("Student name:", s.get_name())
s.set_name("Jane B. Doe")
print("Student name:", s.get_name())
print("courses enrolled in:", s.get_courses())
s.enroll('BTE320')
s.enroll('BTE320')
print(f"Student {s.get_name()} has enrolled in {len(s)} course(s): {s.get_courses()}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    dist = (dx + dy) ** 0.5
    return dist

class Rectangle:
  def __init__(self, x1, y1, x2, y2, x3, y3, x4, y4):
    self.p1 = Point(x1, y1) #upper left
    self.p2 = Point(x2, y2) #upper right
    self.p3 = Point(x3, y3) #lower left
    self.p4 = Point(x4, y4) #lower right
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)
  def area(self):
    return self.width * self.height
  def perimeter(self):
    return
```

```python
class Employee:
  def __init__(self,name,hourly_wage, hours):
    self.name = name
    self.__hourly_wage = hourly_wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self,newWage):
    assert neWage > 0
    self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee name: {self.name}\nHourly charge: {self.__hourly_wage}$"

class Manager(Employee):
  def __init__(self, name, hourly_wage, hours, bonus):
    super().__init__(name, hourly_wage, hours)
    self.bonus = bonus #float between 0-1; eg., 0.2 means 20% extra

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100,15,0.2)
print(employee)
print(manager)
print()
print(f"employee {employee.name} weekly salary: {employee.calculate_pay()}$.")
print(f"manager {manager.name} weekly salary: {manager.calculate_pay()}$ that includes a {manager.bonus *100}% bonus.")
```
